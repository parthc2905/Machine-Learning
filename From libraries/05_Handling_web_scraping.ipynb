```python
soup = BeautifulSoup(response.text , 'lxml')
mydivs = soup.find_all('div', class_="companyCardWrapper")  #class_
mydivs
```

```python
def retReviews(res):
    try:
        return res.find_all('span',class_="companyCardWrapper__ActionCount")[0].text.strip()
    except:
        return None

def retSalaries(res):
    try:
        return res.find_all('span',class_="companyCardWrapper__ActionCount")[1].text.strip()
    except:
        return None

def retInterviews(res):
    try:
        return res.find_all('span',class_="companyCardWrapper__ActionCount")[2].text.strip()
    except:
        return None

def retJobs(res):
    try:
        return res.find_all('span',class_="companyCardWrapper__ActionCount")[3].text.strip()
    except:
        return None

def retBenifits(res):
    try :
        return res.find_all('span',class_="companyCardWrapper__ActionCount")[4].text.strip()
    except:
        return None

def retCompanyName(res):
    try:
        return res.h2.text.strip()
    except:
        return None

def retRating(res):
    try:
        return res.find('div',class_="rating_text").text.strip()
    except:
        None

```

```python
for i in mydivs:
    print(retCompanyName(i),retReviews(i),retRating(i),retSalaries(i),retInterviews(i),retJobs(i),retBenifits(i))
```

```python
rows = []
for i in mydivs:
    rows.append([retCompanyName(i),retReviews(i),retRating(i),retSalaries(i),retInterviews(i),retJobs(i),retBenifits(i)])

rows
```

```python
import pandas as pd

pd.DataFrame(rows,columns=['Company Name','Reviews','Rating','Salaries','Interviews','Jobs','Benifits'])
```

```python
columns=['Company Name','Reviews','Rating','Salaries','Interviews','Jobs','Benifits']
df = pd.DataFrame()
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0'}
for i in range(1,504):
    url = f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={i}"
    response = requests.get(url,headers=headers)
    soup = BeautifulSoup(response.text , 'lxml')
    details = soup.find_all('div', class_="companyCardWrapper")  #class_
    rows = []
    for detail in details:
        rows.append([retCompanyName(detail),retReviews(detail),retRating(detail),retSalaries(detail),retInterviews(detail),retJobs(detail),retBenifits(detail)])
    tempdf = pd.DataFrame(rows, columns=columns)
    df = pd.concat([df,tempdf], ignore_index=True)
    
df.to_csv('../Datasets/companies.csv')
```
