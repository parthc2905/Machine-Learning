```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('../Datasets/train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(titanic,x=titanic['Age'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.scatterplot(tips,x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

2. Bar Plot (Numerical - Categorical)

```python
sns.barplot(titanic,x=titanic['Pclass'],y=titanic['Age'])
```

```python
sns.barplot(titanic,x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(titanic, x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

4. Distplot (Numerical - Categorical)

```python

sns.kdeplot(data=titanic[titanic['Survived']==0], x='Age', label='Not Survived')
sns.kdeplot(data=titanic[titanic['Survived']==1], x='Age', label='Survived')

plt.legend()
```

5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.drop(['PassengerId'],axis=1).groupby(['Embarked']).mean(numeric_only=True)['Survived']*100
```

6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

fig, ax = plt.subplots()
fig.patch.set_facecolor('black')      # Full figure background
ax.patch.set_facecolor('black')       # Axes background
ax.spines['bottom'].set_color('white')  # Border color
ax.spines['top'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['right'].set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')

df.plot(ax=ax)
plt.show()
```
