3. Opening CSV through URL

```python
import requests
from io import StringIO

url = "https://gist.githubusercontent.com/rnirmal/e01acfdaf54a6f9b24e91ba4cae63518/raw/6b589a5c5a851711e20c5eb28f9d54742d1fe2dc/datasets.csv"

response = requests.get(url)
df = pd.read_csv(StringIO(response.text))
df
```

4. Sep parameters

```python
# tsv format file
url = "https://gist.githubusercontent.com/jieter/1a497e1fb7d817100f376f8b58f5d12f/raw/56469c272e2ce113ce538927e111e3d91ffc3c70/data.tsv"

response = requests.get(url)
df = pd.read_csv(StringIO(response.text),sep="\t")
df
```

5. Index_col parameter

```python
# to specify the index column
pd.read_csv("../Datasets/aug_train.csv",index_col="enrollee_id")    
```

6. Header parameter

```python
# if by chance header is present in the data row, we can use header parameter to specify the row number which contains the header
pd.read_csv("../Datasets/aug_train.csv")
```

7. use_col parameter

```python
# if we want use some specific columns from the csv file, we can use usecols parameter
pd.read_csv("../Datasets/aug_train.csv",usecols=['enrollee_id','city_development_index','education_level'])
```

8. Squeeze parameter

```python
# it converts the dataframe to series
pd.read_csv("../Datasets/aug_train.csv", usecols=['education_level']).squeeze()
```

9. Skiprows/nrows Parameter

```python
# if we want to skip some rows from the csv file, we can use skiprows parameter
# if we want to use some specific rows from the csv file, we can use nrows parameter
pd.read_csv("../Datasets/aug_train.csv",skiprows=[0,5])
```

11. Skip bad_lines

```python
# if there are some bad lines in the csv file, we can skip those lines using on_bad_lines parameter
pd.read_csv("../Datasets/aug_train.csv",on_bad_lines='skip')
```

12. Dtypes


```python
# in this we have target column as float, but we want it to be int, so we can use dtype parameter to specify the data type of the columns
pd.read_csv("../Datasets/aug_train.csv",index_col="enrollee_id").info()
```

```python
pd.read_csv("../Datasets/aug_train.csv",index_col='enrollee_id', dtype={'target': 'int'}).info()
```
